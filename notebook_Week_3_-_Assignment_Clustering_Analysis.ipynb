# scrape Wikipedia page using BeautifulSoup

```python
#wikipedia link
wiki = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

#open the page
page = urlopen(wiki)

#soup it up
soup = BeautifulSoup(page)
```

# Wikipedia Scraping - Part 2
 creation of open lists that we can append Postcodes, Boroughs and Neighborhoods to as we scrape our Wikipedia page.  

```python
postcode = []
borough = []
neighborhood = []

table = soup.find("table", { "class" : "wikitable sortable" })
print(table)
```

# Wikipedia Scraping - Part 3

loop over our Wikipedia page and append to our empty lists.   remove the '\n' from our Neighborhoods.  check the lenght of each list to ensure we have an equal number of each.

```python
#scrape the zip codes, boroughs and neighborhoods

for row in table.findAll('tr'):
    cells = row.findAll('td')
    if len(cells)==3:
        postcode.append(cells[0].find(text=True))
        borough.append(cells[1].find(text=True))
        neighborhood.append(cells[2].find(text=True))

neighborhood_2 = []
for i in neighborhood:
    i = i.replace('\n', '')
    neighborhood_2.append(i)

print(postcode)
print(borough)
print(neighborhood_2)

print(len(postcode))
print(len(borough))
print(len(neighborhood_2))
```

# FourSquare - Part 6 

Now let's repeat this process for all of Toronto

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
toronto_venues = getNearbyVenues(names=toronto_df2['Neighborhood'],
                                   latitudes=toronto_df2['Latitude'],
                                   longitudes=toronto_df2['Longitude']
                                  )
```

# Data Prep for Clustering - Part 1

Start preparing our new dataset to run our clustering algos

```python
# one hot encoding
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
toronto_onehot['Neighborhood'] = toronto_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]

neigh = toronto_onehot['Neighborhood']
toronto_onehot.drop(labels=['Neighborhood'], axis=1,inplace = True)
toronto_onehot.insert(0, 'Neighborhood', neigh)

toronto_onehot.head()
```

```python
# And let's examine the new dataframe size.

print(toronto_onehot.shape)
```

# Data Prep for Clustering - Part 2

Group rows by neighborhood and by taking the mean of the frequency of occurrence of each category

```python
toronto_grouped = toronto_onehot.groupby('Neighborhood').mean().reset_index()
toronto_grouped
```

```python
print(toronto_grouped.shape)
```

# Data Prep for Clustering - Part 3
Print each neighborhood along with the top 5 most common venues

```python
num_top_venues = 5

for hood in toronto_grouped['Neighborhood']: #for each neighborhood in Neighborhood
    print("----"+hood+"----") #print these lines + hood name + lines
    temp = toronto_grouped[toronto_grouped['Neighborhood'] == hood].T.reset_index() #temp is Neighborhood, Neighborhood = hood, reshape and reset index
    temp.columns = ['venue','freq'] #new values for temp dataframe are venue and freq
    temp = temp.iloc[1:] #start with rows 1 onward (0 is the column names)
    temp['freq'] = temp['freq'].astype(float) #make sure freq is float
    temp = temp.round({'freq': 2}) #make it two decimals
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues)) #sort by freq and head is 5
    print('\n')
```

# Data Prep for Clustering - Part 4



```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:] #categories are everthing from row 1 on
    row_categories_sorted = row_categories.sort_values(ascending=False) #sort the rows based on values in descending order
    
    return row_categories_sorted.index.values[0:num_top_venues] #return the values indexed by values 0:5
```

# Data Prep for Clustering - Part 5
Create the new dataframe and display the top 10 venue categories for each neighborhood.

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = toronto_grouped['Neighborhood']

for ind in np.arange(toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted
```

# Clustering - Part 1

Run k-means to cluster the neighborhood into 5 clusters.

```python
# set number of clusters
kclusters = 5

toronto_grouped_clustering = toronto_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
toronto_merged = toronto_df2

# add clustering labels
toronto_merged['Cluster Labels'] = kmeans.labels_

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
toronto_merged = toronto_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

toronto_merged # check the last columns!
```

# Clustering - Part 2 

Finally, Visualize the resulting clusters

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i+x+(i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighborhood'], toronto_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
